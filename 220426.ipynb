## Representations Matter

```python
import sklearn
import matplotlib.pyplot as plt

from sklearn import datasets
# Load data
iris= datasets.load_iris()
iris_data = iris['data']
```

```python
plt.scatter(iris_data[:,0], iris_data[:, 1], c=iris_data[:, 3], cmap ='flag')
```

```python
plt.scatter(iris_data[:,2], iris_data[:, 3], c=iris_data[:, 3], cmap ='flag')
```

Proper plotting/visualization shows "easier" separation patterns! Representation can even imply reduction of the number of variables

## Evaluate acuracy

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

# Example
y_true = ["cat", "ant", "cat", "cat", "ant", "bird"]
y_pred = ["ant", "ant", "cat", "cat", "ant", "cat"]
cm = confusion_matrix(y_true, y_pred, labels=["ant", "bird", "cat"])
print(cm)
ConfusionMatrixDisplay.from_predictions(y_true, y_pred, labels=["ant", "bird", "cat"])
```

Confusion matrix is used *after* running an algorithm in order to understand how the accuracy in the results is distributed. Rows: Real values. Columns: Predicted values.
All values in the diagonal are succesful predictions, any numbers outside it are misses.

Although some algorithms might have an overall better acuracy, it's important to keep in mind that confusion matrices give very relevant information about the performance. Maybe some mistakes are inadmissible (e.g. predicting 'bird' for an 'ant' is too dangerous)
